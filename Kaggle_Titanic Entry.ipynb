```python
import numpy as np
import pandas as pd

data=pd.read_csv('test.csv')
```

```python
data.head()

data.info()
data.describe()
```

```python
data.head(20)
```

```python
avg_mean=data.Age.mean()
avg_mean
```

```python
data[['Age']] = data[['Age']].fillna(value=30.272590361445783)
```

```python
data.head(20)
```

```python
pclass = pd.get_dummies(data.Pclass,
                    prefix = 'Pclass')
data = pd.merge(data, pclass,
                    left_index=True, right_index=True)
```

```python
data.Sex.replace(['male','female'],[1,0],inplace=True)
```

```python
test_features = data[['Age', 'Sex', 'Pclass_1',
'Pclass_2', 'Pclass_3']].values 
```

```python
df=pd.read_csv('clean_data.csv')
df.info()
df.describe()
df.head()
```

```python
df_features = df[['Age_all','Sex',
    'Pclass_1','Pclass_2','Pclass_3']].values
df_target = df.Survived.values
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier().fit(df_features,df_target)
```

```python
predictions = model.predict(test_features)
```

```python
data['Survived'] = predictions
```
